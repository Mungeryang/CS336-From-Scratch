bytes 类型是可迭代的,可以像遍历列表一样遍历它，只不过遍历出来的每个元素是该字节对应的整数（0-255）。

函数报错的根本原因在于**UTF-8 是一种变长编码，**字节与字符不总是一一对应的。在 UTF-8 编码中，一个英文字符（ASCII）占用 1 个字节，但一个中文字符通常占用 3 个字节。

bytes([b]).decode("utf-8")意味着试图把每一个独立的字节都当成一个完整的字节去解码。

````python
def decode_utf8_bytes_to_str_wrong(bytestring: bytes):
    return "".join([bytes([b]).decode("utf-8") for b in bytestring])
````

````python
decode_utf8_bytes_to_str_wrong("hello".encode("utf-8"))
````

````python
decode_utf8_bytes_to_str_wrong("你好".encode("utf-8"))
````

UTF-16对基础的拉丁字母强制使用 2 个字节，UTF-32更是强制使用 4 个字节。对于训练语料库来说，代码和英文占比很高，使用UTF-16/32会导致大量的空间浪费(内存中大量的0填充)

b'\xff\xff', 在UTF-8编码规则中，任何字节都不允许以二进制的1111 1111开头，这在UTF-8中是非法的！！！

--> 经验总结：`计算机世界中的反例一般出现在边界值`

### ⚠️为什么要用 regex pretokenizer？

str.split() 会丢失很多重要信息：

```python
" hello\n\nworld".split() # output -> ["hello", "world"]
```

但是在正式的训练中，

- 前导空格很重要

- 连续换行很重要

- 标点怎么和词分开也很重要

所以真正的分词结果应该是：[' hello', '\n\n', 'world']


````python
GPT2_PAT = r"""'(?:[sdmt]|ll|ve|re)| ?\p{L}+| ?\p{N}+| ?[^\s\p{L}\p{N}]+|\s+(?!\S)|\s+"""
````

### 方法抽象

````python
def apply_merge_token(word_bytes_counts,best_pair: tuple):
    merge_token = best_pair[0] + best_pair[1]
    new_word_bytes_counts = Counter()
    
    for word,freq in word_bytes_counts.items():
        new_word = []
        i = 0
        
        while i < len(word):
            if i < len(word) - 1 and (word[i],word[i + 1]) == best_pair:
                new_word.append(merge_token)
                i += 2
            else:
                new_word.append(word[i])
                i += 1
        new_word_bytes_counts[tuple(new_word)] += freq
    return new_word_bytes_counts
````
