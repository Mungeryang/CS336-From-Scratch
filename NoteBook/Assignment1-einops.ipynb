```python
import math
import torch
import torch.nn.functional as F
import torch.nn as nn
from einops import rearrange, einsum, reduce
```

```python
x1 = torch.ones(3, 4) # seq1 hidden
y1 = torch.ones(4, 3) # hidden seq2
```

```python
z1 = x1 @ y1
```

```python
z1
```

```python
zz1 = einsum(x1, y1, "seq1 hidden, hidden seq2 -> seq1 seq2")
```
